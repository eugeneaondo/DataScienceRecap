```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
revenue = pd.read_csv('https://raw.githubusercontent.com/OpenClassrooms-Student-Center/en-8253136-Use-Python-Libraries-for-Data-Science/main/data/CA.csv')
revenue['date'] = pd.to_datetime(revenue['date'])
revenue.head()
```

```python
revenue['date']
```

```python
df = pd.DataFrame(revenue)

# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Group by month and get the count
df.groupby(df['date'].dt.month).size()

# aggregations on other columns
real_estate = df.groupby(df['date'].dt.month)['real estate'].sum()
```

```python
plt.figure(figsize=(10,6))
plt.plot(revenue['date'], revenue['real estate'], label='real estate', linewidth=3)
plt.plot(revenue['date'], revenue['vehicle'], label='vehicle', linewidth=3)
plt.plot(revenue['date'], revenue['consumer'], label='consumer', linewidth=3)
plt.legend()
plt.grid( color= "gray" , linestyle='-', linewidth=0.5)
plt.title("Monthly Net Profit by Month in 2021" )
plt.ylabel("Net Profit($)" , fontsize = 15)
plt.yticks(fontsize = 11)
```
